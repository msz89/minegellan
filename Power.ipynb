```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("POWER.CSV")
```

```python
# df.drop("State", axis=1, inplace = True)
df["Day"] = df['Date'].map(lambda s: s.day)
df["Month"] = df['Date'].map(lambda s: s.month)
df["Year"] = df['Date'].map(lambda s: s.year)
df["Hour"] = df['Date'].map(lambda s: s.hour+s.minute/60)
```

```python
df_j = df[df['Month'] ==1]
```

```python
df_j[df_j["RRP"] > 1000]
# df_j.drop([1375,1376,1377,1472], axis=0, inplace= True)
```

```python
fig = plt.figure()
ax0 = fig.add_subplot(111)
for i in range(1,29): 
    ax0 = df_j[df_j["Day"]==i].plot(kind="line",
                                    x = "Hour",
                                    y="RRP",
                                    ax=ax0,
                                    figsize=(20,10)
                                   )

```

```python
df2 = df_j.groupby("Hour").mean()
df2.reset_index(inplace=True)
df2.plot(kind="line",
        x="Hour",
        y="RRP"
       )
```
